```python

Train test and split
from sklearn.model_selection import train_test_split
```

```python

Removing rows or columns by specifying label names and corresponding axis 'class' is removed in X
X = data.drop('class', axis=1)
```

```python

#labeling the class
Y = data['class']
```

```python
#variable assignment for further study
Q = X
W = Y
E = X
R = Y
A = X
B = Y
C = X
D = Y
```

```python
#code to split
# experimenting the the data split and parameters associated with it such as test_size and random_state
X_train,X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.08, stratify = Y, random_state = 1)
# for random state analysis
Q_train,Q_test, W_train, W_test = train_test_split(Q,W,test_size = 0.08, stratify = W, random_state = 2)
E_train,E_test, R_train, R_test = train_test_split(E,R,test_size = 0.08, stratify = R, random_state = 10)
#for test and train value split analysis
A_train,A_test, B_train, B_test = train_test_split(A,B,test_size = 0.2, stratify = B, random_state = 1)
C_train,C_test, D_train, D_test = train_test_split(C,D,test_size = 0.1, stratify = D, random_state = 1)
```

```python


# Binarization of input X and Y
X_binarised_train = X_train.apply(pd.cut,bins=2, labels=[1,0])
X_binarised_test = X_test.apply(pd.cut,bins=2, labels=[1,0])
type(X_binarised_train)
X_binarised_train = X_binarised_train.values
X_binarised_test = X_binarised_test.values

```

```python
#possible values that b can take
b = 10#possible values 0 - 30
from random import randint
i = randint(0 , X_binarised_train.shape[0])
if(np.sum(X_binarised_train[i,:])>=b):
    print("Model prediction is Malignant")
else:
    print("Model prediction is bengin")   
if (Y_train[i]==1):
    print("Actual prediction is Malignant")
else:
    print("Actual prediction is bengin")
```

```python


#building the model and training it
##b = 10
for b in range(X_binarised_train.shape[1]+1):
    accurate_rows = 0
    for x,y in zip(X_binarised_train,Y_train):
        y_pred = (np.sum(x)>=b)
        accurate_rows +=(y == y_pred)
    print("for b = ",b,accurate_rows,accurate_rows/X_binarised_train.shape[0])
    
    
```

```python



# taking the b value in the test data as we found 28 and 27 to be best accurate in the training.
b = 28
accurate_rows=0
for x,y in zip(X_binarised_test,Y_test):
    y_pred = (np.sum(x)>=b)
    accurate_rows +=(y == y_pred)
    print("for b = ",b,accurate_rows,accurate_rows/X_binarised_test.shape[0])
    
    
    
```
